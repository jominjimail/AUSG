```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
#시그모이드

def sigmoid(x): 
    a = []  
    for itr in x:   
        a.append(1/(1+np.exp(-itr)))  
    return a  

x = np.linspace(-8, 8, 100)  # -8 ~ 8 사이의 100개의 값 생성
sig = sigmoid(x)  
plt.plot(x, sig, color = 'b', marker = "o")  
plt.title("def sigmoid") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.grid(True)
plt.show()  
```

```python
# 하이퍼블릭 탄젠트

def tanh(x): 
    a = []  
    for itr in x:   
        a.append(np.sinh(itr)/np.cosh(itr))  
    return a  

x = np.linspace(-3, 3, 100)  
tan = tanh(x)  
plt.plot(x, tan,color = 'b', marker = "o")
plt.title("def tanh") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.grid(True)
plt.show()  
```

```python
# 렐루

def relu(x): 
    a = []  
    for itr in x:   
        a.append(np.maximum(0,itr))  
    return a  

x = np.linspace(-10, 10, 100) 
ReLu = relu(x)  
plt.plot(x, ReLu,color = 'b', marker = "o")
plt.title("def ReLu") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.grid(True)
plt.show()  
```

```python
# Leaky 렐루

def Leaky_relu(x): 
    a = []  
    for itr in x:   
        a.append(np.where(itr > 0 , itr , itr*0.01))  
    return a  

x = np.linspace(-10, 10, 100)  
L_ReLu = Leaky_relu(x)  
plt.plot(x, L_ReLu,color = 'b', marker = "o")
plt.title("def Leaky_ReLu") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.grid(True)
plt.show()  
```

```python
# 항등 함수

def identity(x): 
    a = []  
    for itr in x:   
        a.append(itr)  
    return a  

x = np.linspace(-10, 10, 100)  
iden = identity(x)  
plt.plot(x, iden,color = 'b', marker = "o")
plt.title("def Identity") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.grid(True)
plt.show()  
```

```python
# 소프트 맥스

def softmax(x): 
    expX = np.exp(x)
    sumExpX = np.sum(expX)
    return expX / sumExpX

x = np.array([2.3 , -0.9 , 3.6])
sof = softmax(x)  

print(sof , np.sum(sof))
```

```python
# 주의 : 지수 함수 계산 결과가 배우 크면 overflos 가 발생할 수 있다.

over = np.array([900 , 1000 , 1000])
over_result = softmax(over)
print(over_result , np.sum(over_result))
```

```python
# 이를 해결하기 위해 sofrmax 함수 구현시 입력 값 중 최대값 찾아 
# (상수 C) 빼주면 올바르게 계산할 수 있다. 

def softmax_sol(x):
    expX = np.exp(x-np.max(x))
    sumExpX = np.sum(expX)
    return expX / sumExpX

over = np.array([900 , 1000 , 1000])
over_result = softmax_sol(over)
print(over_result , np.sum(over_result))
```
